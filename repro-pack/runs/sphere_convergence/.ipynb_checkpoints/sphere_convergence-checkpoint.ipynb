## Mesh refinement study

#### hardware:
- `defq`: 2 * Intel(R) Xeon(R) Gold 6148 CPU @ 2.40GHz, 20 cores/socket, 192GB DRAM

#### mesh: 5PTI

with 5 levels of refinement

| number of elements | element density |
|:------------------:|:---------------:|
|        3,032        |        1        |
|        6,196        |        2        |
|        12,512       |        4        |
|        25,204       |        8        |
|        50,596       |        16       |

#### configuration:
- tolerance = 1e-7
- quadrature order: 4
- fmm expansion order: 10
- fmm ncrit: 500
- restart: 400 (interface from `scipy.sparse.linalg.gmres`, make sure we use full GMRES instead of restarted GMRES)

#### formulation:
- direct formulation: piecewise constant element, block-diagonal preconditioner

$$
\begin{bmatrix}
    \tfrac12\mathrm{I} + \mathrm{K}_L &  -\mathrm{V}_L \\
    \tfrac12\mathrm{I} - \mathrm{K}_H &  \frac{\epsilon_1}{\epsilon_2}\mathrm{V}_H
\end{bmatrix}
\left\{
\begin{matrix}
    \phi \\
    \frac{\partial \phi}{\partial \mathbf{n}}
\end{matrix}
\right\}
= \left\{
\begin{matrix}
    \sum_k \frac{q_k}{4\pi \epsilon_1 \lvert r-r_k \rvert} \\
    0
\end{matrix}
\right\}
$$

- Juffer's formulation: piecewise linear element, mass matrix preconditioner

$$
\begin{bmatrix}
    \frac{1+\epsilon}{2}\mathrm{I} - \epsilon \mathrm{K}_H + \mathrm{K}_L &  \mathrm{V}_H - \mathrm{V}_L \\
    \mathrm{D}_H - \mathrm{D}_L &  \frac{1+1/\epsilon}{2}\mathrm{I} + \frac{1}{\epsilon}\mathrm{T}_H - \mathrm{T}_L
\end{bmatrix}
\left\{
\begin{matrix}
    \phi \\
    \frac{\partial \phi}{\partial \mathbf{n}}
\end{matrix}
\right\}
= \left\{
\begin{matrix}
    \sum_k \frac{q_k}{4\pi \epsilon_1 \lvert \mathbf{r}-\mathbf{r}_k \rvert} \\
    \sum_k \frac{q_k (\mathbf{r}-\mathbf{r}_k) \cdot \mathbf{n}}{4\pi \epsilon_1 \lvert \mathbf{r}-\mathbf{r}_k \rvert^3}
\end{matrix}
\right\}
$$

## Mesh refinement study using a sphere

To verify our BEM-FMM integration, we performed mesh refinement study

```python
direct_sphere_df = get_df('result/sphere_convergence/direct', formulation='direct', skip4=True)
direct_sphere_df
```

```python
juffer_sphere_df = get_df('result/sphere_convergence/juffer', formulation='juffer', skip4=True)
juffer_sphere_df
```

```python
e_solv_exact = -12.2583630
```

```python
e_solv = direct_sphere_df['e_solv [kcal/Mol]'].values
rel_error_direct = np.abs((e_solv-e_solv_exact)/e_solv_exact)
e_solv = juffer_sphere_df['e_solv [kcal/Mol]'].values
rel_error_juffer = np.abs((e_solv-e_solv_exact)/e_solv_exact)
```

```python
rel_error_direct
```

```python
e_solv_collocation = np.array((-12.4326706, -12.301547, -12.269150778, -12.261115, -12.25916835))
rel_error_collocation = np.abs((e_solv_collocation-e_solv_exact)/e_solv_exact)
```

```python
fig = plt.figure(figsize=(3, 2))
ax = fig.add_subplot(111)
N = direct_sphere_df.index.values
N_ = np.array((3e2, 2e5))
asymp = N[2] * rel_error_direct[2] / N_
ax.loglog(N, rel_error_direct, linestyle='', marker='o', fillstyle='none', label='direct', ms=4)
ax.loglog(N, rel_error_juffer, linestyle='', marker='v', fillstyle='none', label='Juffer', ms=4)

ax.loglog(N, rel_error_collocation, linestyle='', marker='x', label='collocation', ms=4)

ax.loglog(N_, asymp, linestyle='--', color='#7f7f7f', linewidth=0.8)
ax.grid(True, which="both")
ax.set_xlabel('number of elements')
ax.set_ylabel('relative error')
ax.legend()

loc = (3*N[-2]+N[-1])/4
text_loc = np.array((loc, 1.2*N[-2]*rel_error_direct[-2]/loc))
ax.text(text_loc[0], text_loc[1],r'N$^{-1}$', fontsize=8, rotation=-30,rotation_mode='anchor');
```

## Mesh Refinement study using 5PTI

### result from using direct formulation

```python
direct_5PTI_df = get_df('result/5PTI_mesh_refinement/direct/', formulation='direct', skip4=True)
direct_5PTI_df
```

### result from using Juffer's formulation

The condition number of the system matrix does not grow with the number of elements.

```python
juffer_5PTI_df = get_df('result/5PTI_mesh_refinement/juffer/', formulation='juffer', skip4=True)
juffer_5PTI_df
```

### Richardson extrapolation

```python
def richardson_extrapolation(f1, f2, f3):
    return (f1*f3 - f2**2) / (f3 - 2*f2 + f1) 
```

```python
e_solv = direct_5PTI_df.e_solv.values
e_solv_exact = richardson_extrapolation(*e_solv[-2:0:-1])
rel_error_direct = np.abs((e_solv-e_solv_exact)/e_solv_exact)
pd.DataFrame({'number of elements': direct_5PTI_df.index, 'relative error (%)': rel_error_direct*100}).set_index('number of elements')
```

```python
e_solv = juffer_5PTI_df.e_solv.values
e_solv_exact = richardson_extrapolation(*e_solv[-2:0:-1])
rel_error_juffer = np.abs((e_solv-e_solv_exact)/e_solv_exact)
pd.DataFrame({'number of elements': direct_5PTI_df.index, 'relative error (%)': rel_error_juffer*100}).set_index('number of elements')
```

```python
fig = plt.figure(figsize=(3, 2))
ax = fig.add_subplot(111)
N = direct_5PTI_df.index.values
N_ = np.array((1e3, 1e5))
asymp = N[2] * rel_error_direct[2] / N_
ax.loglog(N, rel_error_direct, linestyle='', marker='o', fillstyle='none', label='direct')
ax.loglog(N, rel_error_juffer, linestyle='', marker='v', fillstyle='none', label='Juffer')
ax.loglog(N_, asymp, linestyle='--', color='#7f7f7f', linewidth=0.8)
ax.grid(True, which="both")
ax.set_xlabel('number of elements')
ax.set_ylabel('relative error')
ax.legend()

loc = (3*N[-2]+N[-1])/4
text_loc = np.array((loc, 1.2*N[-2]*rel_error_direct[-2]/loc))
ax.text(text_loc[0], text_loc[1],r'N$^{-1}$', fontsize=8, rotation=-30,rotation_mode='anchor');
```
